```python
import numpy as np
import matplotlib.pyplot as plt

from sympy import Symbol, sin, cos, sqrt, lambdify, solve, pi
from sympy.plotting import plot_parametric

from sympy.interactive import printing
printing.init_printing(use_latex=True)

%matplotlib inline
```

```python
t, x, y = Symbol('t'), Symbol('x'), Symbol('y')
solutions = solve([x**2 / 4 + y**2 / 9 - 1, x - 2 + t*y], [x, y]); solutions
```

```python
x0, y0 = solutions[1][0], solutions[1][1]
dx0, dy0 = x0.diff(t), y0.diff(t)
d2x0, d2y0 = dx0.diff(t), dy0.diff(t)
```

```python
x1, y1 = 2*cos(t), 3*sin(t)
dx1, dy1 = x1.diff(t), y1.diff(t)
d2x1, d2y1 = dx1.diff(t), dy1.diff(t)
```

```python
norm_dc0 = sqrt(dx0**2 + dy0**2)
k0 = (dx0 * d2y0 - dy0 * d2x0) / norm_dc0**3

norm_dc1 = sqrt(dx1**2 + dy1**2)
k1 = (dx1 * d2y1 - dy1 * d2x1) / norm_dc1**3
```

```python
k0 = k0.simplify()
k1 = k1.simplify()
```

```python
k0_s = k0.diff(t) / norm_dc0; k0_s = k0_s.simplify()
k1_s = k1.diff(t) / norm_dc1; k1_s = k1_s.simplify();
k0_s, k1_s
```

```python
plot_parametric(k0, k0_s, (t, -10, 10))
```

```python
plot_parametric(k1, k1_s, (t, 0, 2*pi))
```
